# Q6 revisited

```python
#Create a spending flag for each user based on the average price across all their orders using the following criteria:
#If the mean of the prices of products purchased by a user is lower than 10, then flag them as a “Low spender.”
#If the mean of the prices of products purchased by a user is higher than or equal to 10, then flag them as a “High spender.”
```

```python
# Creating “average_price” column using transform()

ords_prods_merge['average_price'] = ords_prods_merge.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
ords_prods_merge.head()
```

```python
# Creating a 'spending_flag' flag

ords_prods_merge.loc[ords_prods_merge['average_price'] < 10, 'spending_flag'] = 'Low spender'
```

```python
# Create 'spending_flag' flag

ords_prods_merge.loc[ords_prods_merge['average_price'] >= 10, 'spending_flag'] = 'High spender'
```

```python
ords_prods_merge.head()
```

```python
ords_prods_merge.shape
```

```python
# Checking frequency
ords_prods_merge['spending_flag'].value_counts(dropna = False)
```

```python
ords_prods_merge.info()
```

```python
# Create “median_last_orders” column using transform()

ords_prods_merge['median_last_orders'] = ords_prods_merge.groupby(['user_id'])['days_since_last_order'].transform(np.median)
```

```python
ords_prods_merge.head()
```

```python
# Create an 'order_frequency_flag' flag

ords_prods_merge.loc[ords_prods_merge['median_last_orders'] > 20, 'order_frequency_flag'] = 'Non-frequent customer'
```

```python
# Creating an 'order_frequency_flag' flag

ords_prods_merge.loc[(ords_prods_merge['median_last_orders'] > 10) & (ords_prods_merge['median_last_orders'] <= 20), 'order_frequency_flag'] = 'Regular customer'
```

```python
# final 'order_frequency_flag' flag

ords_prods_merge.loc[ords_prods_merge['median_last_orders'] <= 10, 'order_frequency_flag'] = 'Frequent customer'
```

```python
ords_prods_merge.head()
```

```python
# Checking the frequency

ords_prods_merge['order_frequency_flag'].value_counts(dropna = False)
```

```python
# Checking the null values

print('Records in ords_prods_merge with null values in column order_frequency_flag:')
ords_prods_merge[ords_prods_merge['order_frequency_flag'].isnull()][['user_id', 'days_since_last_order', 'median_last_orders', 'order_frequency_flag']]
```
